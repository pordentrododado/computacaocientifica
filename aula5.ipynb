É possível resolver sistemas lineares, fazer decomposições de matrizes, entre outras funções.

**(1)** Escreva a função produto_interno que recebe dois vetores de qualquer dimensão e retorna o produto interno desses vetores.

**(2)** Faça uma função que dados a, b e n crie um array com n valores entre a e b (inclusive) igualmente espaçados.

**(3)** Faça uma função que dado um array x retorne um array y com a relação $y=x^2$.

```python
def quad(x):
    return np.array(x)**2
```

```python
quad([1,2,3])
```

```python
quad(np.array([2,3,4]))
```

```python
from PIL import Image
import requests
from io import BytesIO

response = requests.get('https://s3.amazonaws.com/assets.datacamp.com/blog_assets/Machine+Learning+R/iris-machinelearning.png')
Image.open(BytesIO(response.content))
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/siddharthjain1611/Iris_dataset/master/Iris.csv")
df.head()
```

```python
df.shape
```

```python
map_color = {'Iris-setosa':1, 'Iris-versicolor':2, 'Iris-virginica':3}
```

```python
features = ['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']
```

```python
list(map_color.keys())
```

```python
fig = plt.figure(figsize=(20,15))
k = 1

for i in range(len(features)):
    for j in range(len(features)):
        plt.subplot(4,4,k)
        plt.scatter(df[features[i]],df[features[j]], c=df.Species.apply(lambda x: map_color[x]))
        plt.colorbar(ticks=[1,2,3])
        plt.xlabel(features[i])
        plt.ylabel(features[j])
        k=k+1
```
